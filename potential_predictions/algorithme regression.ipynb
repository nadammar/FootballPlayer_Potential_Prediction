```python
import numpy as np
labels=np.array(features['potentiel'])
features=features.drop("potentiel",axis=1)
feature_list=list(features.columns)
features=np.array(features)

```

```python
from sklearn.model_selection import train_test_split
train_features,test_features,train_labels,test_labels = train_test_split(features,labels,test_size=0.25,random_state=42)
print(train_features.shape)
print(train_labels.shape)
print(test_features.shape)
print(test_labels.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=10000,random_state=42)
rf.fit(train_features,train_labels);
predictions=rf.predict(test_features)
errors=abs(predictions-test_labels)
print("mean absolute error:",round(np.mean(errors),2),)
mape=100*(errors/test_labels)
precision=100-np.mean(mape)
print("precision:",round(precision,2),"%.")
```

```python
importances=list(rf.feature_importances_)
importances
```

```python
feature_importances=[(feature,round(importance,2))for feature,importance in zip(feature_list,importances)]
feature_importances=sorted(feature_importances,key=lambda x:x[1],reverse=True)
[print("variable:{:20}importance:{}".format(*pair)) for pair in feature_importances]
```

```python
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
x_values=list(range(len(importances)))
plt.bar(x_values,importances,orientation="vertical")
plt.xticks(x_values,feature_list,rotation="vertical")
plt.ylabel("imporatnce");
plt.xlabel("variable");
plt.title("variable imporatnces");
plt.show()
```

```python
model=RandomForestRegressor()
model=model.fit(train_features,train_labels)
y_pred=model.predict(test_features)
fig,ax=plt.subplots()
ax.plot(test_labels,"b",label="valeur reel")
ax.plot(y_pred,"r",label="valeur predit")
leg=ax.legend();
```
